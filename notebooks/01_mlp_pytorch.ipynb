# Multilayer Perceptron (MLP) with PyTorch

This notebook covers the fundamentals of neural networks and builds up to a working MLP.

**Topics covered:**
- What is a Neuron / Node?
- Perceptron
- Multilayer Perceptron (MLP)
- Activation Functions
- Forward Pass
- Backpropagation & Optimization


## 1. What is a Neuron?

A **neuron** (or node) is the basic unit of a neural network.  
It takes one or more inputs, multiplies each by a **weight**, sums them up, adds a **bias**, then passes the result through an **activation function**.

```
output = activation(w1*x1 + w2*x2 + ... + wn*xn + b)
```

Think of it like a dial: the weights control how much each input matters.


## 2. Perceptron → MLP

A **Perceptron** is a single neuron (the simplest neural network).  
Stack multiple neurons in **layers** and you get a **Multilayer Perceptron (MLP)**:

- **Input layer** — raw data
- **Hidden layer(s)** — learned representations
- **Output layer** — final prediction

The more hidden layers, the deeper the network.


## 3. Activation Functions

Without activation functions, stacking layers is mathematically equivalent to a single linear transformation. Activations introduce **non-linearity**, allowing the network to learn complex patterns.

| Function | Formula | Use case |
|----------|---------|----------|
| ReLU | max(0, x) | Hidden layers (most common) |
| Sigmoid | 1/(1+e^-x) | Binary output |
| Softmax | e^xi / Σe^xj | Multi-class output |


````python
import torch
import torch.nn as nn
import torch.optim as optim

# ── Simple MLP Definition ──────────────────────────────────────────────
class SimpleMLP(nn.Module):
    def __init__(self):
        super().__init__()
        # Hidden layer: 10 inputs → 5 hidden neurons
        self.hidden_layer = nn.Linear(10, 5)
        # Output layer: 5 hidden neurons → 1 output
        self.output_layer = nn.Linear(5, 1)
        # Activation function
        self.relu = nn.ReLU()

    def forward(self, x):
        # FORWARD PASS: data flows left → right
        x = self.hidden_layer(x)
        x = self.relu(x)
        x = self.output_layer(x)
        return x

model = SimpleMLP()
print(model)
print(f"\nTotal parameters: {sum(p.numel() for p in model.parameters()):,}")
````

## 4. Forward Pass, Loss, Backpropagation & Optimization

**Forward pass** — inputs flow through the network layer by layer to produce a prediction.  
**Loss** — measures how wrong the prediction is (e.g. MSE, CrossEntropy).  
**Backpropagation** — computes the gradient of the loss with respect to each weight using the chain rule.  
**Optimization** — uses those gradients to update the weights (e.g. SGD, Adam).


````python
# ── One full training step ────────────────────────────────────────────
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Fake data
dummy_input  = torch.randn(1, 10)   # 1 sample, 10 features
true_answer  = torch.tensor([[1.0]])

# Step 1 — Forward pass (make a prediction)
prediction = model(dummy_input)
print(f"Prediction : {prediction.item():.4f}")
print(f"True answer: {true_answer.item():.4f}")

# Step 2 — Compute loss
loss = loss_function(prediction, true_answer)
print(f"Loss       : {loss.item():.4f}")

# Step 3 — Backpropagation (compute gradients)
loss.backward()

# Step 4 — Update weights
optimizer.step()
optimizer.zero_grad()   # clear gradients for the next step

print("\n✓ One learning cycle complete!")
````
